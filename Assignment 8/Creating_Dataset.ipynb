```python
selected_images = {}
with open("/content/drive/MyDrive/CSULB Assignments/CECS 551 ADV AI/Assignment 8/identity_CelebA.txt") as f:
    for line in f:
        (key, val) = line.split(maxsplit=1)
        val=val.rstrip('\n')
        selected_images[(key)] = val

print(selected_images.items())
```

```python
with open("/content/drive/MyDrive/CSULB Assignments/CECS 551 ADV AI/Assignment 8/selected_ids.txt") as f:

    
    selected_ids = [x.rstrip() for x in f]
    
    print(selected_ids)
```

```python
d = {n:[k for k in selected_images.keys() if selected_images[k] == n] for n in set(selected_images.values())}
print(d)
```

```python
IC=[]

for k,v in d.items():
    
    if(k in selected_ids):
        IC.append(v)
print(IC)
```

```python
len(IC)
```

```python
textfile = open("/content/drive/MyDrive/AI Assignment 8/Selected.txt", "w")
for i in IC:
    for element in i:
        textfile.write(str(element) + "\n")
textfile.close()
```

```python
import os
import shutil

folder_dir = '/content/drive/MyDrive/AI Assignment 8/celebrities.zip'
base_dir = '/content/drive/MyDrive/AI Assignment 8/base_dir'

os.makedirs(base_dir, exist_ok=True)  

filenames = open("/content/drive/MyDrive/AI Assignment 8/Selected.txt").read().split('\n')

for name in filenames:
    if name:  
        fullpath = os.path.join(folder_dir, name)
        if os.path.exists(fullpath):
            print('coping:', fullpath)
            shutil.copy(fullpath, base_dir)
        else:
            print('Transferring', fullpath)
```
