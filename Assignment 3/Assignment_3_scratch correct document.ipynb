```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import make_moons,make_circles,make_classification
```

```python
lr = 0.02
epochs = 10000
```

```python
#Sigmoid function

def sigmoid(x):
    s = 1/(1+np.exp(-x))
    return s
```

```python
# Binary cross entropy as loss function

def loss(y,p):
    l = -(np.mean(y*np.log(p) + (1-y)*np.log(1-p))) 
    print(l)
    return l   
```

```python
#Initialising weights and biases
np.random.seed(0)  cxd  
W1 = np.random.randn(2,2)
b1 = np.random.randn(1,2)
W2 = np.random.randn(2,1)
b2 = np.random.randn(1,1)
```

```python
#initialising weights

X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])
```

```python
#Forward Propagation
def forward(X):
  z1 = np.dot(X,W1) + b1
  a1 = np.tanh(z1) 
  z2 = np.dot(a1,W2) + b2
  y_hat = sigmoid(z2)

  forward_value = (a1,y_hat)

  return forward_value
```

```python
#Backward Propagation

def backward(x, y, y_hat, a1, W1, b1, W2, b2):
  del2 = y_hat - y
  dw2 = np.dot(a1.T,del2) 
  db2 = np.sum(del2,axis=0)
  
  del1 = (1-np.square(a1))*np.dot(del2,W2.T)  
  dw1 = np.dot(x.T,del1)  
  db1 = np.sum(del1,axis=0)
  
  W1 -= lr*dw1
  W2 -= lr*dw2
  b1 -= lr*db1
  b2 -= lr*db2

  backward_value = (W1, b1, W2, b2)

  return backward_value
```

```python
training_loss = []
val_loss = []
val_acc = []
for i in range(epochs):
  a1, y_pred = forward(X)
  l = loss(y,y_pred)
  output = backward(X, y, y_pred, a1, W1, b1, W2, b2)
  
  print("Epoch ", + i + "Loss ", + l)# %f"%(i,l),end='\n')

#model.summary()
```
